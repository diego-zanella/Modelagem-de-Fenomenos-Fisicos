1- Escreva uma função em Python que compute o valor da função abaixo, emitindo uma mensagem de erro caso a função receba um valor não permitido. Além disso, a função deverá plotar um gráfico da função no intervalo entre ±10, destacando neste gráfico os pontos relacionados ao valor de passado para a função..


$ f(x\big) =  \frac{1}{(\sqrt{2 - x})}$

```python
import matplotlib.pyplot as plt
import numpy as np

def ex_01(x):
    if x >= 2:
      print("Valor de \"x\" inválido, função não calculada.")
      return None
    else:
      res = 1.0 / np.sqrt(2 - x)
      print(f"\nResultado da função: {res}")
      x_vals = np.linspace(-10, 1.9, 1000)
      y_vals = np.zeros_like(x_vals) # cria um vetor de zeros com a mesma dimensão de x_vals
      mask = 2 - x_vals > 0 # cria uma máscara booleana para selecionar apenas os valores de x para os quais a raiz quadrada de 2 - x é positiva
      y_vals[mask] = 1.0 / np.sqrt(2 - x_vals[mask]) # calcula os valores da função apenas para os valores de x selecionados pela máscara
      plt.plot(x_vals, y_vals, 'b', label='Função')
      plt.xlabel('x')
      plt.ylabel('f(x)')
      plt.title('Gráfico da função f(x)')
      plt.plot(x, 1.0 / np.sqrt(2 - x), 'ro')
      plt.xlim([-10, 10])
      plt.legend()
      plt.show()

x = float(input("Insira o valor de \"x\" para calcular a função: "))
ex_01(x)
```

2 - Utilizando a linguagem Python, e somente as bibliotecas Matplotlib e Numpy, crie uma
função para plotar uma raquete de tênis de mesa. No seu gráfico a raquete deve ser plotada
de forma que seja possível observar tanto o cabo quanto a área de rebatimento.


```python
import matplotlib.pyplot as plt

def plota_raquete():
    fig, ax = plt.subplots()

    # círculo
    raio = 1.0
    circulo = plt.Circle((0, 0), raio, facecolor='#FF0000', edgecolor='black', linewidth=1)
    ax.add_artist(circulo)

    # retângulo
    largura, altura = 0.4, raio
    retangulo = plt.Rectangle((-0.2, -2), largura, altura, facecolor='#DECD8B', edgecolor='black', linewidth=1)
    ax.add_artist(retangulo)

    # gráfico
    plt.xlim(-4, 4)
    plt.ylim(-4, 3)
    plt.axis('equal')
    plt.title('Raquete de Tênis de Mesa')
    plt.show()


plota_raquete()
```

3- Utilizando a linguagem Python, nenhuma biblioteca extra e List Comprehension crie uma
função que devolva o resultado da fórmula a seguir para 𝑛 elementos.

$f(x) = 4\left(1 - \sum\limits_{1}^{n} \frac{1}{1+2n}\right)$

```python
def calcular_formula(n):
    soma = 0
    for i in range(1, n+1):
      soma += 1/(1+2 * i)
    resultado = 4 * (1 - soma)
    return resultado

n = int(input("Insira o valor de n: "))
resultado = calcular_formula(n)
print(resultado)
```

4 - Usando a linguagem Python, sem utilizar as bibliotecas Sympy e Numpy, escreva uma função
que receba um valor 𝑛 no domínio dos números reais e devolva o limite da função dada a
seguir quando 𝑥 tende a 𝑛.
$ f(x\big) =  \frac{3x^2 - 1}{(\sqrt{x})}$

```python
import math

def limite(n):
    # Verifica se o valor de n é menor que ou igual a 0.
    if n <= 0:
      print("Limite não pode ser calculado")
      return None

    # Define a função f(x).
    def f(x):
        return (3 * x ** 2 - 1) / math.sqrt(x)

    # Calcula o limite de f(x) quando x tende a n.
    if n > 0:
        return f(n)
    else:
        return None

n = int(input("Insira o valor de n: "))
limite(n)
```

5 - ClockCom é uma fábrica de relógios digitais capaz de produzir 1500 relógios por mês. Quando
eles vendem 𝑥 relógios em um mês seu lucro é dado por

$ f(x\big) =  300000 - 36000x + 75x^2 - \frac{x^3}{3}$

Quantos relógios eles precisam vender por mês para maximizar seu lucro? Além de responder
a pergunta referente ao lucro máximo, você precisará fazer uma função que devolva o lucro
obtido para um determinado número 𝑥 de relógios vendidos em um determinado mês e
plote esta função, destacando o ponto de lucro máximo e o ponto referente a quantidade
vendida em um mês específico.

-------------------
**Resposta:**

Para maximizar o lucro, precisamos encontrar o valor de x que maximiza a função f(x). Para isso, podemos calcular a derivada da função f(x) em relação a x, igualá-la a zero e resolver para x:

$f(x) = 300000 - 36000x + 75x^2 - \frac{x^3}{3}$

$f'(x) = -36000 + 150x - x^2$

Igualando a derivada a zero, temos:

$-36000 + 150x - x^2 = 0$

$x^2 - 150x + 36000 = 0$

Resolvendo a equação acima, temos:

$x = \frac{150 \pm \sqrt{150^2 - 4(1)(36000)}}{2(1)} = \frac{150 \pm 300}{2} = 75, 225$

Portanto, a fábrica precisa vender 75 ou 225 relógios por mês para maximizar seu lucro.


```python
import numpy as np
import matplotlib.pyplot as plt

def calcular_lucro(x):
    return 300000 - 36000*x + 75*(x**2) - (x**3)/3

quantidades = np.arange(0, 1500)
lucros = [calcular_lucro(x) for x in quantidades]

plt.plot(quantidades, lucros)
plt.xlabel('Quantidade de relógios vendidos por mês')
plt.ylabel('Lucro (R$)')
plt.title('Lucro da ClockCom em função da quantidade de relógios vendidos')

x_max_lucro = 75
lucro_max = calcular_lucro(x_max_lucro)
plt.plot(x_max_lucro, lucro_max, 'ro')
plt.annotate('Lucro máximo', xy=(x_max_lucro, lucro_max), xytext=(x_max_lucro - 150, lucro_max + 5000),
             arrowprops=dict(facecolor='black', shrink=0.05))

x_mes_especifico = 800
lucro_mes_especifico = calcular_lucro(x_mes_especifico)
plt.plot(x_mes_especifico, lucro_mes_especifico, 'bo')
plt.annotate('Vendido em um mês específico', xy=(x_mes_especifico, lucro_mes_especifico),
             xytext=(x_mes_especifico - 200, lucro_mes_especifico - 20000),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
